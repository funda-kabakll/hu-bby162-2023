```python
dosya = "/content/drive/MyDrive/Colab Notebooks/veri.txt"



def menu():
  print("___Katalog Kayıtlarına Hoşgeldiniz___")
  print("  Menü => " + "1) Kayıtları listele" + "\n" +  "          2) Yeni kayıt ekle" + "\n" + "          3) Tüm kayıtları temizle *Dikkat!"  "\n" +
        "          4) Kayıt sil" + "\n" + "          5) Kayıtlarda arama yap(eser adına göre)" + "\n" + "          6) Güncelleme yap" + "\n" + 
        "          -) Çıkış yap")
  sec = input("Yapmak istediğiniz işlemi seçiniz(1/2/3/4/5/6/-) => ")
  if sec == "-":
    print("Program sonlandırılıyor...!")
  elif sec == "1":
    listele()
  elif sec == "2":
    ekle()
  elif sec == "3":
    temizle()
  elif sec == "4":
    sil()
  elif sec == "5":
    ara()
  elif sec == "6":
    guncelle() 
  else:
    print("Lütfen bu öğelerden birini seçiniz (1/2/3/4/5/6/-) =>  ")
    menu()


def guncelle():
  islem = open(dosya, "r", encoding="utf-8")
  kayitlar = islem.readlines()
  islem.close()

  print("Mevcut kitaplar:")
  for satir in kayitlar:
    kitap = satir.strip()
    ayir = kitap.split("-")
    kitapİsmi = ayir[0]
    print(kitapİsmi)
  guncellenecek = input("Güncellenecek kitap ismini giriniz: ")

  islem = open(dosya, "w", encoding="utf-8")
  for satir in kayitlar:
    kitap = satir.strip()
    ayir = kitap.split("-")
    kitapİsmi = ayir[0]
    if kitapİsmi == guncellenecek.capitalize():
      yeniİsim = input("Yeni kitap ismini giriniz: ")
      yeniYazar = input("Yeni yazarın adını giriniz: ")
      yeniBasim = input("Yeni basım yılını giriniz: ")
      kayit = yeniİsim.capitalize() + "-" + yeniYazar.capitalize() + "-" + yeniBasim
    islem.write(kayit + "\n")
  islem.close()
  menu()


def ara():
  islem = open(dosya, "r", encoding="utf-8")
  aranacak = input("Aranacak kitap ismini giriniz: ")
  for satir in islem:
    kitap = satir.strip()
    ayir = kitap.split("-")
    kitapİsmi = ayir[0]
    if kitapİsmi == aranacak.capitalize():
      print("Aramanızla eşleşen kayıtlar: " + satir)
  islem.close()
  menu()


def sil():
  islem = open(dosya, "r", encoding="utf-8")
  kayitlar = islem.readlines()
  islem.close()

  islem = open(dosya, "w", encoding="utf-8")
  silinecek = (input("Silinecek kitap ismini giriniz: ")).capitalize()
  for satir in kayitlar:
    kitap = satir.strip()
    ayir = kitap.split("-")
    kitapİsmi = ayir[0]
    if kitapİsmi != silinecek:
        islem.write(satir)
  islem.close()
  menu()


def temizle():
  islem = open(dosya, "w")
  islem.close()
  menu()


def ekle():
  islem = open(dosya, "a", encoding = "utf-8")
  isim = input("Kitap ismini giriniz: ")
  yazar = input("Yazarın adını giriniz: ")
  basim = input("Basım yılını giriniz: ")
  islem.write(isim.capitalize() + "-" + yazar.capitalize() + "-" + basim + "\n" )
  islem.close()
  menu()


def listele():
  islem = open(dosya, "r")
  for satir in islem:
    print(satir.strip()) 
  islem.close()
  menu()


menu()
```
