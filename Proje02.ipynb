```python
# Adam Asmaca Oyunu
# 3 fonksiyonun da tek farkı can sayıları ama daha kısa şekilde yapamadım, tekrara düştü
def basit():
  from random import choice
  kelimeler = ["bilgi", "sembol", "data", "bilgelik", "belge"]
  secilenKelime = choice(kelimeler)
# Sözlük şeklinde yapılamadı. Bunun yerine hepsini ayrı ayrı tanımladım
  if secilenKelime == "bilgi": 
    print("İnsan zekâsının çalışması sonucu ortaya çıkan düşünce ürünü, malumat, vukuf.")
  elif secilenKelime == "sembol":
    print("Eş anlamı 'simge' olan kelime.")
  elif secilenKelime == "data":
    print("Eş anlamı 'veri' olan kelime.")
  elif secilenKelime == "bilgelik":
    print("Bilge olma durumu ve niteliği.")
  elif secilenKelime == "belge":
    print("Bir gerçeğe tanıklık eden yazı, fotoğraf, resim, film vb., vesika, doküman.")
  dizilenKelime = []
  for diz in secilenKelime:
    dizilenKelime.append("_")
  print(dizilenKelime)

  adamCan = 5
  while adamCan > 0:
      girilenHarf = input("Bir Harf Giriniz: ")
      canKontrol = girilenHarf in secilenKelime
      if canKontrol == False:
          adamCan-=1
          print("Maalesef Harf Kelimenin İçinde Yok.")
          print("Kalan Can: "+ str(adamCan))
      i = 0
      for kontrol in secilenKelime:
          if secilenKelime[i] == girilenHarf:
              dizilenKelime[i] = girilenHarf
              print("İşte Bu, Aynen Devam!")
          i+=1
      print(dizilenKelime)
  if adamCan == 0:
    print("Maalesef Canınız Bitti :(")
  
# Can tükenmeden doğru tahmin edilse bile oyun sonlanmıyor. Bu yüzden son tahmini sordum
  kelimeOmu = input("Tahmin Edilen Kelimeyi Yazınız =") 
  if kelimeOmu == secilenKelime:
    print("Bravo, Doğru Tahmin : ")
  else:
    print("Maalesef, Tahmininiz Yanlış :( ")

def orta():
  from random import choice
  kelimeler = ["bilgi", "sembol", "data", "bilgelik", "belge"]
  secilenKelime = choice(kelimeler)
  if secilenKelime == "bilgi":
    print("İnsan zekâsının çalışması sonucu ortaya çıkan düşünce ürünü, malumat, vukuf.")
  elif secilenKelime == "sembol":
    print("Eş anlamı 'simge' olan kelime.")
  elif secilenKelime == "data":
    print("Eş anlamı 'veri' olan kelime.")
  elif secilenKelime == "bilgelik":
    print("Bilge olma durumu ve niteliği.")
  elif secilenKelime == "belge":
    print("Bir gerçeğe tanıklık eden yazı, fotoğraf, resim, film vb., vesika, doküman.")
  dizilenKelime = []
  for diz in secilenKelime:
    dizilenKelime.append("_")
  print(dizilenKelime)

  adamCan = 3
  while adamCan > 0:
      girilenHarf = input("Bir Harf Giriniz: ")
      canKontrol = girilenHarf in secilenKelime
      if canKontrol == False:
          adamCan-=1
          print("Maalesef Harf Kelimenin İçinde Yok.")
          print("Kalan Can: "+ str(adamCan))
      i = 0
      for kontrol in secilenKelime:
          if secilenKelime[i] == girilenHarf:
              dizilenKelime[i] = girilenHarf
              print("İşte Bu, Aynen Devam!")
          i+=1
      print(dizilenKelime)
  if adamCan == 0:
    print("Maalesef Canınız Bitti :(")

  kelimeOmu = input("Tahmin Edilen Kelimeyi Yazınız =")
  if kelimeOmu == secilenKelime:
    print("Bravo, Doğru Tahmin :) ")
  else:
    print("Maalesef, Tahmininiz Yanlış :( ")

def ileri():
  from random import choice
  kelimeler = ["bilgi", "sembol", "data", "bilgelik", "belge"]
  secilenKelime = choice(kelimeler)
  if secilenKelime == "bilgi":
    print("İnsan zekâsının çalışması sonucu ortaya çıkan düşünce ürünü, malumat, vukuf.")
  elif secilenKelime == "sembol":
    print("Eş anlamı 'simge' olan kelime.")
  elif secilenKelime == "data":
    print("Eş anlamı 'veri' olan kelime.")
  elif secilenKelime == "bilgelik":
    print("Bilge olma durumu ve niteliği.")
  elif secilenKelime == "belge":
    print("Bir gerçeğe tanıklık eden yazı, fotoğraf, resim, film vb., vesika, doküman.")
  dizilenKelime = []
  for diz in secilenKelime:
    dizilenKelime.append("_")
  print(dizilenKelime)

  adamCan = 1
  while adamCan > 0:
      girilenHarf = input("Bir Harf Giriniz: ")
      canKontrol = girilenHarf in secilenKelime
      if canKontrol == False:
          adamCan-=1
          print("Maalesef Harf Kelimenin İçinde Yok.")
          print("Kalan Can: "+ str(adamCan))
      i = 0
      for kontrol in secilenKelime:
          if secilenKelime[i] == girilenHarf:
              dizilenKelime[i] = girilenHarf
              print("İşte Bu, Aynen Devam!")
          i+=1
      print(dizilenKelime)
  if adamCan == 0:
    print("Maalesef Canınız Bitti.")

  kelimeOmu = input("Tahmin Edilen Kelimeyi Yazınız = ")
  if kelimeOmu == secilenKelime:
    print("Bravo, Doğru Tahmin :) ")
  else:
    print("Maalesef, Tahmininiz Yanlış :( ")



def adam():
  print(" ^_^   Adam Asmaca Oyununa Hoş Geldiniz   ^_^ ")
  seviye = input("Oynamak İstediğiniz Seviyeyi Seçiniz = Basit(b) / Orta(o) / İleri(i) ").lower()
  if seviye == "basit" or seviye == "b" :
    print("___ Hazırsanız Başlıyoruz ___")
    basit()
  elif seviye == "orta" or seviye == "o" :
    print("___ Hazırsanız Başlıyoruz ___")
    orta()
  elif seviye == "ileri" or seviye == "i" :
    print("___ Hazırsanız Başlıyoruz ___")
    ileri()
  else:
    print("Lütfen Geçerli Bir Seviye Seçiniz..")
    adam()

adam()

while True:
  tekrar = input("Tekrar Oynamak İster Misiniz? (Evet/Hayır) ").lower()
  if tekrar == "e" or tekrar == "evet" :
    adam()
  else:
    print("Program Sonlandırılıyor... Oynamayı unutma, yine gel! ^_^ ")
    break
```
